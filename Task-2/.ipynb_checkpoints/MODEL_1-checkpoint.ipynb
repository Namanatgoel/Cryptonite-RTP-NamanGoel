# POKEMON DATASET K-MEANS CLUSTERING AND DBSCAN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("pokemon.csv")  # <-- change path if needed
print(df.head())
```

```python
# Numerical features for clustering
features = ["hp", "attack", "defense", "sp_attack", "sp_defense", "speed", "height_m","speed","weight_kg"]
X = df[features].dropna().values

# Normalization
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

```python
# PCA
X_centered = X - X.mean(axis=0)
cov_matrix = np.cov(X_centered, rowvar=False)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort eigenvectors by eigenvalues descending
sorted_idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_idx]
eigenvectors = eigenvectors[:, sorted_idx]

PC1 = eigenvectors[:, 0]
PC2 = eigenvectors[:, 1]

# Feature contributions
loading_df = pd.DataFrame({'Feature': features, 'PC1': PC1, 'PC2': PC2})
print("\nFeature contributions to PC1 and PC2:")
print(loading_df)

# Top 2 features
top_pc1_feature = loading_df.loc[np.abs(loading_df['PC1']).idxmax(), 'Feature']
top_pc2_feature = loading_df.loc[np.abs(loading_df['PC2']).idxmax(), 'Feature']
print(f"\nMost important feature for PC1: {top_pc1_feature}")
print(f"Most important feature for PC2: {top_pc2_feature}")

# Transform to 2D PCA space
X_pca = X_centered @ eigenvectors[:, :2]

# Variance explained
var_explained = eigenvalues / np.sum(eigenvalues)
print("\nVariance explained by each PC:")
for i, v in enumerate(var_explained[:5]):
    print(f"PC{i+1}: {v:.2f}")
```

```python
# K-Means from scratch
def kmeans(X, k, max_iters=100):
    np.random.seed(42) #seed value fixed here to reproduce the same result but for an actual and more indepth analysis, we will make this also loop randomly till we find the best one
    centroids = X[np.random.choice(X.shape[0], k, replace=False)]
    for _ in range(max_iters):
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)
        new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(k)])
        if np.allclose(centroids, new_centroids):
            break
        centroids = new_centroids
    return labels, centroids
# Elbow method
def compute_wcss(X, k):
    labels, centroids = kmeans(X, k)
    return sum(np.sum((X[labels == j] - centroids[j])**2) for j in range(k))

K_range = range(1, 10)
wcss_values = [compute_wcss(X, k) for k in K_range]

plt.figure(figsize=(7,5))
plt.plot(K_range, wcss_values, 'bo-')
plt.xlabel("Number of clusters k")
plt.ylabel("WCSS")
plt.title("Elbow Method (from scratch)")
plt.show()

# Run K-Means
k = 3  # chosen from elbow
kmeans_labels, kmeans_centroids = kmeans(X, k)
print(f"\nK-Means cluster counts: {np.bincount(kmeans_labels)}")
```

```python
# DBSCAN from scratch
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))

def region_query(X, point_idx, eps):
    neighbors = []
    for i in range(X.shape[0]):
        if euclidean_distance(X[point_idx], X[i]) <= eps:
            neighbors.append(i)
    return neighbors

def expand_cluster(X, labels, point_idx, cluster_id, eps, min_pts):
    neighbors = region_query(X, point_idx, eps)
    if len(neighbors) < min_pts:
        labels[point_idx] = -1  # noise
        return False
    else:
        labels[point_idx] = cluster_id
        i = 0
        while i < len(neighbors):
            neighbor_idx = neighbors[i]
            if labels[neighbor_idx] == -1:
                labels[neighbor_idx] = cluster_id
            elif labels[neighbor_idx] == 0:
                labels[neighbor_idx] = cluster_id
                new_neighbors = region_query(X, neighbor_idx, eps)
                if len(new_neighbors) >= min_pts:
                    neighbors += new_neighbors
            i += 1
        return True

def dbscan(X, eps, min_pts):
    cluster_id = 0
    labels = np.zeros(X.shape[0], dtype=int)
    for point_idx in range(X.shape[0]):
        if labels[point_idx] != 0:
            continue
        if expand_cluster(X, labels, point_idx, cluster_id + 1, eps, min_pts):
            cluster_id += 1
    return labels

eps = 0.8
min_pts = 5
dbscan_labels = dbscan(X_pca, eps=eps, min_pts=min_pts)
print("\nDBSCAN unique cluster labels:", np.unique(dbscan_labels))
print(f"Total clusters found (excluding noise): {len(np.unique(dbscan_labels[dbscan_labels > 0]))}")
print(f"Noise points: {np.sum(dbscan_labels == -1)}")
```

```python
# PCA Biplot (K-Means + DBSCAN)
plt.figure(figsize=(16,7))

# Left: K-Means
plt.subplot(1,2,1)
scatter = plt.scatter(X_pca[:,0], X_pca[:,1], c=kmeans_labels, cmap='viridis', s=50, alpha=0.6)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("K-Means Clusters")

# Centroids
centroids_pca = (kmeans_centroids - X.mean(axis=0)) @ eigenvectors[:, :2]
plt.scatter(centroids_pca[:,0], centroids_pca[:,1], color='red', marker='X', s=200, label='Centroids')
plt.grid(True)
plt.legend()

# Right: DBSCAN
plt.subplot(1,2,2)
scatter2 = plt.scatter(X_pca[:,0], X_pca[:,1], c=dbscan_labels, cmap='tab10', s=50, alpha=0.6)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("DBSCAN Clusters")
plt.grid(True)

plt.show()
```

```python
# PCA biplot arrows (features) on K-Means plot
plt.figure(figsize=(10,7))
plt.scatter(X_pca[:,0], X_pca[:,1], c=kmeans_labels, cmap='viridis', s=50, alpha=0.6)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("K-Means Clusters with Feature Arrows")
# Draw feature arrows
for i, feature in enumerate(features):
    plt.arrow(0,0, PC1[i]*4, PC2[i]*4, color='r', alpha=0.8, head_width=0.15)
    plt.text(PC1[i]*4.3, PC2[i]*4.3, feature, color='r', fontsize=10)
# Highlight top features
plt.arrow(0,0, PC1[features.index(top_pc1_feature)]*5, PC2[features.index(top_pc1_feature)]*5,
          color='blue', alpha=1, head_width=0.2)
plt.text(PC1[features.index(top_pc1_feature)]*5.3, PC2[features.index(top_pc1_feature)]*5.3,
         f"Top PC1: {top_pc1_feature}", color='blue', fontsize=11)
plt.arrow(0,0, PC1[features.index(top_pc2_feature)]*5, PC2[features.index(top_pc2_feature)]*5,
          color='green', alpha=1, head_width=0.2)
plt.text(PC1[features.index(top_pc2_feature)]*5.3, PC2[features.index(top_pc2_feature)]*5.3,
         f"Top PC2: {top_pc2_feature}", color='green', fontsize=11)
plt.grid(True)
plt.show()
```

**CONCLUSION**

Clusters found: 3 (based on elbow method).

Only 2 PC's taken for they contribute the maximum variation in this dataset

Cluster characteristics:  
Cluster 1  Big/powerful Pokmon (high PC1: height/weight/attack)  
Cluster 2  Small/weaker Pokmon (low PC1)  
Cluster 3  Fast Pokmon (high PC2: speed)  

PC1 (height/weight/attack) separates big vs small Pokmon.  
PC2 (speed) separates fast vs slow Pokmon.  
K-Means clusters align well with this 2D PCA visualization.  
DBSCAN clusters may identify the same groups but also label rare Pokmon as noise.  

For this Pokmon dataset,   
K-Means is more suitable because the clusters are mostly spherical in the normalized 8D stats space, we have a clear idea of 3 clusters, and we want interpretable centroids.  
DBSCAN is useful if we expect irregular clusters or want to identify outlier Pokmon, but here it has only grouped the data into 1 and -1 (for outlier pokemon's) so it is not useful.
